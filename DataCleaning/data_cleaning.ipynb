```python
import pandas as pd
pokeDex = pd.read_csv("../input/pokemon/Pokemon.csv")
print(pokeDex.to_string())
pokeDex = pokeDex.to_numpy()
```

```python
onehotType = np.zeros((pokeDex.shape[0],18),dtype=np.int8) #18 types
typeList = ["Fire", "Water", "Grass", "Electric", "Psychic", "Steel", "Normal", "Fairy", "Dark", "Flying", "Ghost", "Poison", "Ice", "Ground", "Rock", "Dragon", "Fighting","Bug"]
```

```python
print(typeList.index("Grass"))
print("From {} types".format(len(typeList)))
```

```python
type1 = pokeDex[:,2]
type2 = pokeDex[:,4]
names = pokeDex[:,1]
```

```python
for i in range(pokeDex.shape[0]):
    i_type1 = type1[i]
    type1_index = typeList.index(i_type1)
    onehotType[i,type1_index] = 1
    assert(np.sum(onehotType[i,0:])>0)
    
```

```python
df = {'name': names}
for i in range(len(typeList)):
    typeName = typeList[i]
    typeShot = onehotType[:,i]
    df[typeName] = typeShot
df = pd.DataFrame(df)
```
